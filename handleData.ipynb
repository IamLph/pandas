```python
import pandas as pd
```

```python
# 读取数据
fpath = "./data/test2.csv"
data = pd.read_csv(fpath)
data
```

```python
# 替换后缀cc
data.loc[:,"sum"] = data["sum"].str.replace("cc","").astype("float")
data
```

```python
# Series排序 升序
data["sum"].sort_values()
```

```python
# dataframe排序
data.sort_values(by="sum")
# 降序
data.sort_values(by="sum",ascending=False)
```

```python
# 多列排序 从左到右 按照升序排序
data.sort_values(by=["regin","sum"])
```

```python
# 多列 降序
data.sort_values(by=["regin","sum"],ascending=False)
```

```python
# 分别指定升序和降序
data.sort_values(by=["regin","sum"],ascending=[True,False])
```
